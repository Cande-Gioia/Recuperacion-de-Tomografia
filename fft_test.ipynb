```python
from skimage import color, io, util, transform
from matplotlib import pyplot as plt
```

```python
image = io.imread('gabu.jpg')
image = image[...,0]
```

```python
plt.figure()
plt.imshow(image, cmap='gray')
```

```python
from numpy.fft import fft2, ifftshift, fftshift
import numpy as np
```

```python
transform = fftshift(fft2(ifftshift(image)))
```

```python
plt.figure()
plt.imshow(abs(transform), vmin = 0.0, vmax = 1.0)#, cmap='grey')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-500, 501, 1)
X, Y = np.meshgrid(x, x)
wavelength = 100
angle = 0
grating = np.sin(
    2*np.pi*(X*np.cos(angle) + Y*np.sin(angle)) / wavelength
)
plt.set_cmap("gray")
plt.subplot(121)
plt.imshow(grating)
# Calculate Fourier transform of grating
ft = np.fft.ifftshift(grating)
ft = np.fft.fft2(ft)
ft = np.fft.fftshift(ft)
plt.subplot(122)
plt.imshow(abs(ft))
plt.xlim([480, 520])
plt.ylim([520, 480])  # Note, order is reversed for y
plt.show()
```
